#### Solution:

To evaluate the fine-tuned Bert model is used for sentiment analysis we need to consider following evaluation matrics:
- Common metrics used for evaluation include accuracy, precision, recall, and F1 score.
-  Accuracy can be a suitable metric as the cost of a false positive  or a false negative  is approximately equal.
- Precision can be a suitable evaluation metric, particularly when the cost of false positives is high.
- When the goal of a prediction task is to minimize false negatives, recall serves as an appropriate evaluation metric.
- The F1 score is an appropriate evaluation metric, as the goal is to strike a balance between precision and recall.

#### Solution:

To evaluate the Neural Machine Translation Model we need to consider following evaluation matrics:

- LEPOR is a combination of many evaluation factors including existing ones (precision, recall) and modified ones (sentence-length penalty and n-gram based word order penalty).

##### Problem
- Let's consider a scenario involving a fine-tuned LLM that's designed for entity extraction, specifically targeting companies and persons within text data. However, the model is showing signs of overfitting, as it's struggling to correctly identify and distinguish between persons and companies.



- Your task is to figure out/propose solutions to overcome overfitting in this situation.

- You've taken a base pre-trained language model and fine-tuned it for entity extraction. The goal is to identify and classify persons and companies within text documents. However, as you test the fine-tuned model on various text samples, you observe that the model frequently misclassifies entities.

- Signs of Overfitting

   - The model's predictions are overly confident but incorrect when identifying persons and companies.
   - It struggles to generalize to new data, particularly when confronted with variations in text structure, word usage, and entity context.

- Performing Data augmentation
- Feature selection
- Using Cross-validation i.e. split dataset into parts.Then, we iteratively train the algorithm on-1 folds while using the remaining holdout fold as the test set.This method allows us to tune the hyperparameters of the neural network or machine learning model and test it using completely unseen data.
- Using Regularization methods like Lasso, L1 can be beneficial if we do not know which features to remove from our model.
- Early stopping during the training phase (have an eye over the loss over the training period as soon as loss begins to increase stop training).

#### Exercise 4

- Catastrophic Forgetting in a Fine-Tuned Language Model for Text Classification
- Problem
  - In this scenario, we'll explore how a pre-trained Large Language Model (LLM) can face catastrophic forgetting when fine-tuned for a new text classification task, causing it to lose its ability to accurately classify certain types of text that it was proficient at before fine-tuning.

- Your task is to figure out how to adjust the model to overcome this situation?
- Scenario Description
  - Imagine a pre-trained LLM that has been extensively trained on a wide variety of text documents, spanning different domains such as news, social media posts, wikipedia etc. The LLM is capable of performing sentiment analysis, topic categorization and other text classification tasks with high accuracy.
  - However, when you decide to fine-tune the LLM specifically for categorizing scientific articles, its performance in classifying other types of text, such as movie reviews or news articles, deteriorates significantly.

- Signs of Catastrophic Forgetting

  - The LLM's classification accuracy drops dramatically for tasks it was proficient at before fine-tuning.
  - Despite initially being successful at classifying diverse text types, the fine-tuned LLM provides incorrect or inconsistent classifications for those text categories.
  - The LLM becomes proficient at categorizing scientific articles but loses its versatility for other text classification tasks.




#### Solution:

To Overcome the Catastropic forgetting we can use the following techniques:
   - EWC(Elastic Weight Consolidation) which is a method that tries to find the overlap of low error regions for task A and task B by slowing down the learning on weights that are most important for task A while training on task B.
   - Multi-task Finetuning: Progressive Learning
        - Initially, the model focuses on pre-training knowledge and slowly incorporates the new task data, minimizing the risk of catastrophic forgetting.
   - Parameter Efficient Finetuning: Transfer Learning
     - Freeze certain layers of the model during fine-tuning.
     - By freezing early layers responsible for fundamental language understanding, we preserve the core knowledge while only fine-tuning later layers for the specific task.
